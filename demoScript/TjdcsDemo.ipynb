## tjdcs库的使用方法 

```python
# 滤波器对比
from tjdcs import sFilter
import numpy as np
import matplotlib.pyplot as plt
fil10 = sFilter(Bs = [1.0], As = [10, 1])
fil20 = sFilter(Bs = [1.0], As = [20, 1])

N1, N2 = 100, 100
u = np.hstack([np.random.randn(N1), 3*np.ones(N2)])
y10 = []
y20 = []

for k in range(0,N1+N2):
    yk10 = fil10(u[k], y_ini=u[k])
    yk20 = fil20(u[k], y_ini=u[k])
    
    y10.append(yk10)
    y20.append(yk20)

plt.plot(u)
plt.plot(y10, label='y10')
plt.plot(y20, label='y20')
plt.legend()
plt.show()  
```
